```python
cat = ['RECORD', 'BBLE', 'B', 'BLOCK', 'LOT', 'EASEMENT', 'OWNER', 'BLDGCL',
       'TAXCLASS', 'EXT', 'EXCD1', 'STADDR', 'ZIP', 'EXMPTCL', 'EXCD2', 
       'PERIOD', 'YEAR', 'VALTYPE']
num = ['LTFRONT', 'LTDEPTH', 'STORIES', 'FULLVAL', 'AVLAND', 'AVTOT', 
       'EXLAND', 'EXTOT', 'BLDFRONT', 'BLDDEPTH', 'AVLAND2', 
       'AVTOT2', 'EXLAND2', 'EXTOT2',]
```

```python
df.columns
```

```python
num_tbl = df[num].describe().transpose()
```

```python
num_tbl = num_tbl.rename(columns={'count':'# Records', 'std':'Standard Deviation'})
```

```python
num_tbl = num_tbl.drop(['25%','50%','75%'], axis=1)
```

```python
num_tbl['% Populated'] = num_tbl['# Records']*100/1070994
```

```python
num_tbl
```

```python
for i in num_tbl.index:
    num_tbl.loc[i, '# Zeros'] = sum(df[i] == 0)
# sum(df.LTFRONT == 0)
```

```python
num_tbl = num_tbl.reset_index()
```

```python
num_tbl = num_tbl.rename(columns = {'index':'Field Name'})
```

```python
num_tbl['Field Type'] = 'Numeric'
```

```python
num_tbl.round(2)
```

```python
cat_tbl = pd.DataFrame(cat)
```

```python
cat_tbl = cat_tbl.rename(columns = {0:'Field Name'})
```

```python
cat_tbl['# Records'] = list(1070994-df[i].isnull().sum() for i in cat)
```

```python
cat_tbl['% Populated'] = cat_tbl['# Records']/1070994
```

```python
cat_tbl['# Unique'] = 0
```

```python
for i in range(18):
    cat_tbl.iloc[i,3] = len(df[cat[i]].unique())

```

```python
cat_tbl['Most Common'] = list(df[i].value_counts().index[0] for i in cat)
```

```python
for i in range(18):
    if cat_tbl.iloc[i,3] == 1070994:
        cat_tbl.iloc[i,4] = 'N/A'
```
